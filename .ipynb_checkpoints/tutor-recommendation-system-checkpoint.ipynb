# Introduction

In this notebook, a Tutor Recommendation System is built using the provided documents in the email for the assessment which include  `students.pdf` and `tutors.pdf` files, the methological approach employed is a combination of a content base filtering with scoring system for feature engineering and  LightGBM (Gradient Boost Model) and ANN (Artificial Neural Network) model  for recommendation and ranking. 

# Why LightGBM and ANN
LightGBM was choosen because it is  optimized for speed and memory usage which makes it suitable for relatively small datasets like this one, It natively supports ranking tasks which aligns perfectly with the recommendation systemâ€™s goal of ranking tutors for each student also LightGBM can efficiently handle categorical variables after encoding, ensuring strong performance on structured tabular data another reason is that feature importance scores from LightGBM provide insights into which student-tutor attributes drive recommendations.

Initially for this project, I only want to use LightGBM, but the choice of ANN came because it allows  exploration of deep learning approaches for recommendation, which can scale to larger datasets in the future. ANN can capture complex, non-linear patterns between student preferences and tutor attributes that may not be obvious in tabular features. Neural networks adapt well to diverse feature types and can learn intricate interactions beyond simple scoring rules also ANN provides a complementary perspective to LightGBM.

# The methodology steps carried out in brief:
1. Data Preprocessing : In this step, the given dataset was converted from pdf file format to pandas DataFrame datatype which is machine learning friendly using `camelot-py` library,  data cleaning is not carried out because the dataset is short and there is no missing/null/nan values.
2. Feature Engineering: The content base filtering approach uses a scoring system to include additional features (column) named `target_label` that capture the similarities between students and tutors, this method generate the column by filtering different content base logical and idea conditions such as if the student's hobby intercept with tutor's hobby, another example is a case in-which the student's prefered programming language matches the tutor's language and so on. By applying content-based method, the process creates context awareness and pattern for the models to learn and adapt with by giving a `target_label` which serves as a rank feature for each observation in the dataset.
3. Encoding Categorical or Dummy variables: In this step each columns containing categorical or dummy variables were handled using `LabelEncoder` from  `sklearn.preprocessing` module. Additionally, `hobbies` column in both students and tutors dataset was tokenized and converted into set datatype, this is carried out because each observation in the students `hobbies` column was use to filter intersection with tutors `columns`.
4. Feature Seperation: After succesful data preprocessing and feature engineering, features and target label were extracted.
5. Data Split: The dataset is then split into train set and test set use 80% and 20% respectively, then each set was grouped according to `student_id`, this ensure that the dataset is ready for LightGBM and ANN model training.
6. Model Training: During this phase LightGBM and ANN model was trained using the train set and test set from 5 above.
7. Model Evaluation and Testing: The Performance of the LightGBM  and ANN model was evaluted using metrics such as normalized discounted cummalative gain `(NDCG@k)` and discounted cummulative gain `(dcg)`, the higher metrics the more accurate prediction for the models,  these two metrics are then used to calculate the relevance score of each tutor to student, the higher this relevance score the more similarity and recommendation.

# Implementing Feature Engineering Using Scoring System

It is good to note that a pre-existing assumption is feed into the score system in order to generate the target features that the model will use for recommendation, for example in a real-life scenario a beginner will prefer a teaching style that is either structured, interactive or visual, then an intermediate student might prefer hands-on, conversational , project based work or interactive and an advance student might prefer concepts that are advance, analytical or project based, all these concepts are total assumptions that sounds reasonable for a recommendation system.

## How the Scoring System Work

The score system uses 3 features to compute similarities which are languages, choice of programing language and prefered learning or teaching style, each of these 3 features is assigned with weights respectively which are `3,4,2`.  and on each observervation the total score is then use as the `target_label`.

```python
# teaching style preference and assumptions
# this dictionary map each assumptions stated above 
# this is used to represent each assumptions
level_style_preference = {
    "beginner": {"structured", "interactive", "visual"},
    "intermediate": {"hands-on", "conversational", "project-based", "interactive"},
    "advanced": {"advanced-concepts", "analytical", "project-based"}
}

language_weight = 3
programming_weight = 4
level_style_weight = 2

def scoring_system(student_row, tutor_row):
    score = 0
    if student_row["learning_language"].lower() == tutor_row["expertise_languages"].lower():
        score += language_weight
    if student_row["programming_focus"].lower() == tutor_row["programming_specialty"].lower():
        score += programming_weight
    score += len(student_row['hobbies'].intersection(tutor_row['hobbies']))
    if tutor_row["teaching_style"].lower() in level_style_preference[student_row["level"]]:
        score += level_style_weight
    return score
```

```python
# to build the target columns both students and tutors dataset is 
# loop through and the score is calculated using s = student row and t = tutor row
# as argument passed into the scoring_system to build the target_label
pair_rows = []
for _, s in students.iterrows():
    for _, t in tutors.iterrows():
        pair_rows.append({
            "student_id" : s['student_id'], # this is kept for grouping
            "tutor_id": t['tutor_id'], # this is kept for grouping and will be deleted later.
            "is_language_match": int(s["learning_language"].lower() == t["expertise_languages"].lower()),
            "is_programming_match": int(s["programming_focus"].lower() == t["programming_specialty"].lower()),
            "does_hobby_overlap": len(s['hobbies'].intersection(t['hobbies'])),
            "is_teaching_style_match": int(t["teaching_style"].lower() in level_style_preference[s["level"]]),
             "student_level": s["level"],
            "student_prog": s["programming_focus"],
            "tutor_prog": t["programming_specialty"],
            "target_label": scoring_system(s, t)   # integer relevance label
        })

engineered_features = pd.DataFrame(pair_rows)
# save the engineered features
engineered_features.to_csv("engineered_features.csv")
```

# Model Development/Traning

```python
# converting the grouped data into lgb Dataset datatype to use to train LightGBM Model
lgb_train = lgb.Dataset(X_train, label=y_train, group=group_train)
lgb_val   = lgb.Dataset(X_test,  label=y_test,  group=group_test, reference=lgb_train)
lgb_params = {
    "objective": "lambdarank", 
    "metric": "ndcg",
    "ndcg_eval_at": [1, 3],   # only monitor NDCG@1 and NDCG@3 that is k=3
    "learning_rate": 0.1,
    "num_leaves": 10,
    "min_data_in_leaf": 1,
    "verbosity": 3,
    "seed": 42, # for consistent output
    "force_row_wise": True
}

lightgbm_model = lgb.train(
        lgb_params, 
        lgb_train, 
        valid_sets=[lgb_train, lgb_val], 
        valid_names=['train','valid'],
        num_boost_round=5
)

# saving the LightGBM model
lightgbm_model.save_model("lambdarank_model.txt")
```

# Evaluating Model Performance using NDCG@K metric


```python
# calculating the discounted cummulative gain
# using log2* (position+1)  formular
# this ensures that higher-ranked items are less discounted
def dcg_at_k(rels, k):
    rels = np.asarray(rels)[:k]
    if rels.size == 0:
        return 0.0
    discounts = np.log2(np.arange(2, rels.size + 2))
    return np.sum((2**rels - 1) / discounts)

# calculating the normalize discounted cummulative gain
def ndcg_at_k(true_rels, pred_scores, k):
    # sort the predicted scores in descending order
    order = np.argsort(pred_scores)[::-1]
    # get the ideal scores base on the sorted order
    sorted_true = np.asarray(true_rels)[order]
    ideal_sorted = np.sort(true_rels)[::-1]
    idcg = dcg_at_k(ideal_sorted, k)
    if idcg == 0:
        return 0.0
    return dcg_at_k(sorted_true, k) / idcg

def evaluate_by_student(model_predict_fn, df_group, k=3):
    rows = []
    for sid in sorted(df_group['student_id'].unique()):
        # group students by student_id
        group = df_group[df_group['student_id'] == sid]
        Xg = group[features].values
        y_true = group['target_label'].values
        # predict the relvance 
        y_pred = model_predict_fn(Xg)
        # get the normalize discounted cummulative gain
        ndcg = ndcg_at_k(y_true, y_pred, k)
        rows.append({"student_id": sid, f"ndcg@{k}": ndcg})
    return pd.DataFrame(rows)


def get_recommendation_for_students(student_df , model_pred_fn):
    all_recommendations = []
    # LightGBM recs:
    for sid in sorted(student_df['student_id'].unique()):
        # first group the students base on their id
        group = student_df[student_df['student_id'] == sid]

        # get recommendation for top 3 that is k=3 students 
        recommendation = topk_for_student(model_pred_fn, group, k=3)
        all_recommendations.append(recommendation)
    return np.array(all_recommendations)

rec_columns = ['student_id','tutor_id','pred_relevance_score','target_label']
def topk_for_student(model_predict_fn, df_group, k=3):
    Xg = df_group[features].values
    preds = model_predict_fn(Xg)
    df_group = df_group.copy()
    df_group['pred_relevance_score'] = preds
    df_group = df_group.sort_values('pred_relevance_score', ascending=False)
    return df_group.head(k)[rec_columns]

def convert_to_dataframe(recommendations):
    recommendations = recommendations.reshape(-1, recommendations.shape[-1]) 
    recommendations = pd.DataFrame(recommendations, columns=rec_columns)
    print(f"Recommendation Before shape is {recommendations.shape}")
    

    # now add tutor name and student name
    recommendations = recommendations.merge(
    students.rename(columns={'name': 'student_name' , "hobbies": "student_hobbies" , "learning_language":"student_language"}),
    on='student_id',
    how='left'
    )
    
    # merge tutor names
    
    recommendations = recommendations.merge(
        tutors.rename(columns={'name': 'tutor_name' , "hobbies": "tutor_hobbies" , "expertise_languages": "tutor_language" }),
        on='tutor_id',
        how='left'
    )


    print(f"Recommendation shape is {recommendations.shape}")
    
    return recommendations
```

```python
# LightGBM predictions wrapper
lgb_pred_fn = lambda X: lightgbm_model.predict(X, num_iteration=lightgbm_model.best_iteration)
lgb_eval = evaluate_by_student(lgb_pred_fn, test_df, k=3)
lgb_mean_ndcg3 = lgb_eval[f"ndcg@3"].mean()
```

```python
lgb_summary = {
    "lambdarank_ndcg_by_student": lgb_eval.to_dict(orient='records'),
    "lambdarank_mean_ndcg@3": float(lgb_mean_ndcg3),
}

print(lgb_summary)
# saving summary for future use
with open("evaluation_summary.json", "w") as f:
    json.dump(lgb_summary, f, indent=2)

# visualizing evalution summary
lgb_students = [entry['student_id'] for entry in lgb_summary['lambdarank_ndcg_by_student']]
lgb_scores = [entry['ndcg@3'] for entry in lgb_summary['lambdarank_ndcg_by_student']]
lgb_mean_score = lgb_summary['lambdarank_mean_ndcg@3']

# plotting 
plt.figure(figsize=(7,5))
plt.bar(lgb_students, lgb_scores, color=['skyblue','orange'])
plt.axhline(lgb_mean_score, color='red', linestyle='--', label=f'Mean NDCG@3 = {lgb_mean_score:.3f}')
plt.title('LightGBM NDCG@3 Scores per Student')
plt.ylabel('NDCG@3 (0â€“1)')
plt.ylim(0,1.05)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
